Task 1: Exploring and Visualizing a Simple Dataset

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the dataset
df = pd.read_csv("Iris.csv")
```

```python
# Show structure
print("Shape:", df.shape)
```

```python
print("Columns:", df.columns)
```

```python
df.head(15)
```

```python
# Set visual theme
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)  # Default size
```

```python
#Scatter plot
plt.figure()
sns.scatterplot(
    data=df,
    x="SepalLengthCm", 
    y="PetalLengthCm",
    hue="Species",
    palette="Set2",
    s=100,  # Bubble size
    edgecolor='black'
)
plt.title("Sepal vs Petal Length  Separated by Species\n", fontsize=16, fontweight='bold')
plt.xlabel("\nSepal Length (cm)", fontsize=12, fontweight='bold')
plt.ylabel("Petal Length (cm)\n", fontsize=12, fontweight='bold')
plt.legend(title="Species", loc='upper left')
plt.grid(True)
plt.show()
```

```python
# Histogram
plt.figure()
sns.histplot(
    data=df,
    x="SepalWidthCm",
    kde=True,
    bins=20,
    color='skyblue',
    edgecolor='black'
)
plt.title("Distribution of Sepal Width\n", fontsize=16, fontweight='bold')
plt.xlabel("\nSepal Width (cm)", fontsize=12, fontweight='bold')
plt.ylabel("Count\n", fontsize=12, fontweight='bold')
plt.grid(True)
plt.show()
```

```python
# Box plot
plt.figure()
sns.boxplot(
    data=df,
    x="Species",
    y="PetalWidthCm"
)
plt.title("Petal Width Comparison Across Species\n", fontsize=16, fontweight='bold')
plt.xlabel("\nSpecies",fontsize=12, fontweight='bold')
plt.ylabel("Petal Width (cm)\n", fontsize=12, fontweight='bold')
plt.show()
```
